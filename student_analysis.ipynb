**Student Performance Analysis**

```python
import pandas as pd

# Load the datasets
d1 = pd.read_csv(r"D:\meena\student-mat.csv")
d2 = pd.read_csv(r"D:\meena\studentpor.csv")

# Merge the datasets on the given columns
d3 = pd.merge(
    d1,
    d2,
    on=[
        "school", "sex", "age", "address", "famsize", "Pstatus",
        "Medu", "Fedu", "Mjob", "Fjob", "reason", "nursery", "internet"
    ]
)

# Print number of rows
print(len(d3),"students")  # should give 382 students
```

**Step : 3 Analysis Questions**

```python
average_g3_math = d3["G3_x"].mean()
average_g3_por = d3["G3_y"].mean()

print("Average Final Grade (g3_x) = ", average_g3_math)
print("Average Final Grade (g3_y) = ", average_g3_por)
```

```python
above_math = (d3["G3_x"] > 15).sum()
above_por = (d3["G3_y"] > 15).sum()

print("Students scoring above 15 in Mat = ", above_math)
print("Students scoring above 15 in Por = ", above_por)
```

```python
corr_math = d3["studytime_x"].corr(d3["G3_x"])
print("Correlation (studytime vs performance in Mat) =", corr_math)

corr_por = d3["studytime_y"].corr(d3["G3_y"])
print("Correlation (studytime vs performance in Por) =", corr_por)

```

```python
gen_per_math = d3.groupby("sex")["G3_x"].mean()
gen_per_por = d3.groupby("sex")["G3_y"].mean()

print("Average performance by gender (Mat):\n", gen_per_math)
print("\nAverage performance by gender (Port):\n", gen_per_por)
```
