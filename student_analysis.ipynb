**Student Performance Analysis**

**Step : 1 Load the Dataset**

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#load dataset
data = pd.read_csv(r"D:\meena\student-mat.csv")
df = pd.DataFrame(data)
df.head()
```

**Step : 2 Explore and Clean Data**

```python
df.info()
```

```python
#check missing values
print("null values :\n",df.isnull().sum())
```

```python
#remove duplicates
print("Duplicates :",df.duplicated().sum())
```

```python
#shape and types of data
print("Shape :",df.shape)
print("Data Types :\n",df.dtypes)
```

**Step : 3 Analysis Questions**

```python
# Average final grade (g3)
average_g3 = df["G3"].mean()
print("Average Final Grade (G3) = ",average_g3)
```

```python
# how many students scored above 15
above = (df["G3"]>15) . sum()
print("Sum of students who have scored above 15 = ",above)
```

```python
#correlation of studytime and performance
Correlation = df["studytime"].corr(df["G3"])
print("correlation = ",Correlation)
```

```python
#which gender performs better on average
gen_per = df.groupby("sex")["G3"].mean()
print("Gender performance on average \n",gen_per)
```

**Step : 4 Visualization**

```python
df.columns
```

```python
#Histogram
plt.figure(figsize=(8,6))
sns.histplot(df["G1"],kde=True,bins = 10,color="pink")
plt.title("Histogram of Grade(G1)")
plt.xlabel("G1")
plt.ylabel("Number of students")
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8,6))
sns.histplot(df["G2"],kde=True,bins = 10,color="green")
plt.title("Histogram of Grade(G2)")
plt.xlabel("G2")
plt.ylabel("Number of students")
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8,6))
sns.histplot(df["G3"],kde=True,bins = 10,color="yellow")
plt.title("Histogram of Grade(G3)")
plt.xlabel("G3")
plt.ylabel("Number of students")
plt.show()
```

```python
#Scatterplot
plt.figure(figsize=(8,6))
plt.scatter(df['studytime'],df['G3'],color="red")
plt.title("Scatterplot of studytime and final grade")
plt.xlabel('Studytime')
plt.ylabel('G3')
plt.show()
```

```python
#Average final grade by gender
plt.figure(figsize=(8,6))
sns.barplot(x = "sex", y = "G3", data = df, palette="pastel")
plt.title("Average final grade by gender")
plt.xlabel("Gender")
plt.ylabel("Final Grade")
plt.show()
```
